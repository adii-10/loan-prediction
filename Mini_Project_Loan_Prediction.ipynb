```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
import joblib
```

```python
df=pd.read_csv("loan.csv")
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.isnull().sum()*100/len(df)
```

```python
df=df.drop('Loan_ID',axis=1)
```

```python
column=['Gender','Dependents','LoanAmount','Loan_Amount_Term']
```

```python
df=df.dropna(subset=column)
```

```python
df.isnull().sum()*100/len(df)
```

```python
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
df.isnull().sum()*100/len(df)
```

```python
df['Credit_History'].unique()
```

```python
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
df.isnull().sum()*100/len(df)
```

```python
df.sample(10)
```

```python
df['Dependents']=df['Dependents'].replace(to_replace="3+",value='4')
```

```python
df.sample(10)
```

```python
df['Gender']=df['Gender'].map({'Male':1,'Female':0}).astype('int')
df['Married']=df['Married'].map({'Yes':1,'No':0}).astype('int')
df['Education']=df['Education'].map({'Graduate':1,'Not Graduate':0}).astype('int')
df['Self_Employed']=df['Self_Employed'].map({'Yes':1,'No':0}).astype('int')
df['Property_Area']=df['Property_Area'].map({'Urban':1,'Rural':0,'Semiurban':2}).astype('int')
df['Loan_Status']=df['Loan_Status'].map({'Y':1,'N':0}).astype('int')
```

```python
df.head()
```

```python
X=df.drop('Loan_Status',axis=1)
```

```python
Y=df['Loan_Status']
```

```python
c=['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']
```

```python
st =StandardScaler()
X[c]=st.fit_transform(X[c])
```

```python
X
```

```python
model_df={}
def model_value(model,X,Y):
    X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=42)
    model.fit(X_train,Y_train)
    Y_pred=model.predict(X_test)
    print(f"{model} accuracy is {accuracy_score(Y_test,Y_pred)}")
    score=cross_val_score(model,X,Y,cv=5)
    print(f"{model} Avg Cross value score is {np.mean(score)}")
    model_df[model]=round(np.mean(score)*100,2)
```

```python
model=LogisticRegression()
model_value(model,X,Y)
model_df
```

```python
model=svm.SVC()
model_value(model,X,Y)
model_df
```

```python
model=DecisionTreeClassifier()
model_value(model,X,Y)
model_df
```

```python
model=RandomForestClassifier()
model_value(model,X,Y)
model_df
```

```python
model=GradientBoostingClassifier()
model_value(model,X,Y)
model_df
```

```python
log_reg_grid={"C":np.logspace(-4,4,20),"solver":["liblinear"]}
```

```python
rs_log_reg=RandomizedSearchCV(LogisticRegression(),param_distributions=log_reg_grid,n_iter=20,cv=5,verbose=True)
```

```python
rs_log_reg.fit(X,Y)
```

```python
rs_log_reg.best_score_
```

```python
rs_log_reg.best_params_
```

```python
rs_svc=RandomizedSearchCV(svm.SVC(),param_distributions=svc_grid,cv=5,n_iter=20,verbose=True)
```

```python
rs_svc.fit(X,Y)
```

```python
rs_svc.best_score_
```

```python
rs_svc.best_params_
```
